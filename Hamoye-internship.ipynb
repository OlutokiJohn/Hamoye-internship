```python
#reading the data from the csv file
data = pd.read_csv("https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv")
```

```python
# Solving for the standard deviation and 75th percentile of the measure of energy per unit (Fuel_mmbtu_per_unit) in 
# two decimal places?
data.describe()
```

```python
# What is the skewness for the fuel quantity burned in two decimal places?
data.skew(axis = 0, skipna = True )
```

```python
# What is the kurtosis for the fuel quantity burned in two decimal places?
data.kurt(axis = 0, skipna = True )
```

```python
# Which of the following fuel type code has the lowest average fuel cost per unit burned? Coal
data.groupby('fuel_type_code_pudl').agg({'fuel_cost_per_unit_burned': ['mean']})
```
