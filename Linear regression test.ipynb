```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
import requests 
from sklearn.preprocessing import MinMaxScaler #for normalization
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split 
```

```python
data = pd.read_csv('energydata_complete.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().any()
```

```python
data.describe()
```

```python
#removing the following columns: [“date”, “lights”]
data_new = data.drop(columns=['lights','date'])
```

```python
data_new.info()
```

```python
data_new.head()
```

```python
#Doing normalization with Minmax scaling 
scaler = MinMaxScaler()
data_normalized = pd.DataFrame(scaler.fit_transform(data_new), columns=data_new.columns)
```

```python
#cheacking out the normalized datas

data_normalized.head()
```

```python
#defining the datas into attributes(variables) and labels(target)

data_attr = data_normalized.drop(columns = ['Appliances'])
appliances_label = data_normalized['Appliances']
```

```python
data_attr.head()
```

```python
appliances_label.head()
```

<h2>Train Test Split</h2>

```python
# splitting the data

x_train, x_test, y_train, y_test = train_test_split(data_attr, appliances_label, test_size = 0.3, random_state = 42)
```

<h2>Linear Regression</h2>

```python
#fit the model to training dataset
LR = LinearRegression()

LR.fit(x_train, y_train)

#obtain predictions
y_predicted = LR.predict(x_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_predicted))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_predicted))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_predicted)))
print('R2 score', metrics.r2_score(y_test, y_predicted))
```

Question: From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). What is the R^2 value in two D.P? 

```python
x = data['T2']
y = data['T6']
```

```python
x = np.array(x).reshape(-1, 1)
```

```python
y = np.array(y).reshape(-1, 1)
```

```python
# linear model on the relationship between the temperature in the living room in Celsius (x = T2) 
#and the temperature outside the building (y = T6)

# creating an object of LinearRegression class
LR = LinearRegression()

LR.fit(x, y)


#R^2 value in two D.P
r_sq = LR.score(x, y)

print('coefficient of determination:', r_sq)
```

<h2>Ridge Regression</h2>

```python
#perfoming Ridge regression with alpha of 0.5
Ridge_regression = Ridge(alpha = 0.5)

Ridge_regression.fit(x_train, y_train)
ridge_prediction = Ridge_regression.predict(x_test)

#using the model to predict the test data 
print('Root Mean Square (RMSE)', np.sqrt(metrics.mean_squared_error(y_test,ridge_prediction))) #RMSE
print('R2 Score', metrics.r2_score(y_test,ridge_prediction)) #RMSE
```

```python
#perfoming Ridge regression with alpha of 2
Ridge_regression_2 = Ridge(alpha = 2)

Ridge_regression_2.fit(x_train, y_train)
ridge_prediction_2 = Ridge_regression_2.predict(x_test)

#using the model to predict the test data 
print('Root Mean Square (RMSE)', np.sqrt(metrics.mean_squared_error(y_test,ridge_prediction_2))) #RMSE
print('R2 Score', metrics.r2_score(y_test,ridge_prediction_2)) #RMSE
```

```python
#perfoming Ridge regression with alpha of 4
Ridge_regression_3 = Ridge(alpha = 4)

Ridge_regression_3.fit(x_train, y_train)
ridge_prediction_3 = Ridge_regression_3.predict(x_test)

#using the model to predict the test data 
print('Root Mean Square (RMSE)', np.sqrt(metrics.mean_squared_error(y_test,ridge_prediction_3))) #RMSE
print('R2 Score', metrics.r2_score(y_test,ridge_prediction_3)) #RMSE
```

<h1>LASSO REGRESSION </h2>

```python
#perfoming lASSO regression with alpha of 0.001
Lasso_regression = Lasso(alpha = 0.001)

Lasso_regression.fit(x_train, y_train)
lasso_prediction = Lasso_regression.predict(x_test)

#using the model to predict the test data 
print('Root Mean Square (RMSE)', np.sqrt(metrics.mean_squared_error(y_test,lasso_prediction))) #RMSE
print('R2 Score', metrics.r2_score(y_test,lasso_prediction)) #RMSE
```
